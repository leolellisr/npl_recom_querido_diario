```python
def tensor2np(tuple):
  return (np.concatenate([t.numpy() for t in tuple]))

def calc_corr(A,B):
  A = A.reshape(A.shape[0]*A.shape[1]*A.shape[2])
  B = B.reshape(B.shape[0]*B.shape[1]*B.shape[2])
  return np.corrcoef(A,B)
```

```python
import pandas as pd

def knn(matrix,k):
    abc = pd.DataFrame(matrix)
    avg = 0
    for i in range (len(matrix)):
        ret = abc[i].nlargest(k+1)
        toL = ret.values.tolist()
        aux =0
        for j in range(1,k+1):
            aux +=toL[j]
        aux = aux/(k)
        avg+= aux
    return avg/len(matrix)
```
