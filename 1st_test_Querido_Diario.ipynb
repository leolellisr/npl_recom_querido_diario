```python
# URLLib para baixar PDFs
import urllib.request as urllib2

# SSL para sempre permitir download
import ssl
```

```python
# PyPDF para ler PDFs
from PyPDF2 import PdfFileReader
```

```python
# Baixar, extrair informações e textos dos PDFs

infos = []
corpus = []
bad_url = []
i_ok = 0
i_err = 0

#f extract_information(pdf_path):
for pdf_path in paths:
    context = ssl.create_default_context()
    context.check_hostname = False
    context.verify_mode = ssl.CERT_NONE
    try:
      response = urllib2.urlopen(pdf_path, context=context)    
      file = open("temp.pdf", 'wb')
      file.write(response.read())
      allText = []
      with open("temp.pdf", 'rb') as f:
          pdf = PdfFileReader(f)
          information = pdf.getDocumentInfo()
          number_of_pages = pdf.getNumPages()
          for page in range(number_of_pages):
            allText.append(pdf.getPage(page).extractText())
      info = f"""
      Information about {pdf_path}: 

      Author: {information.author}
      Creator: {information.creator}
      Producer: {information.producer}
      Subject: {information.subject}
      Title: {information.title}
      Number of pages: {number_of_pages}
      """

      infos.append(information)
      corpus.extend(allText)
      i_ok += 1
    except:
      bad_url.append(pdf_path)
      i_err += 1
    print(f'done: {i_ok}, error: {i_err}')
```

```python
# Tokenizando com Bertimbau

def my_tokenizer(text):
    # get tokens
    list_tokens = [token.lower() for token in re.compile('\w+').findall(text)]
    
    # clean tokens
    cleaned_tokenizer = [word for word in list_tokens if not word in stop_words_pt]
    
    # convert tokens -> string again
    cleaned_string = ' '.join(cleaned_tokenizer)

    # get bertimbau tokens
    tokens_bert = bertinbau_tokenizer.tokenize(cleaned_string)
    
    # Divide tokens em conjuntos de < 512 
    to_return = []
    size = 512
    arr = tokens_bert
    while len(tokens_bert) > size:
      pice = arr[:size]
      to_return.append(pice)
      arr   = arr[size:]
    to_return.append(arr)         
    return to_return
```

```python
# Preprocessing
# Agradecimentos ao Fábio Collado que fez a maioria das funções de pré-processamento

def remove_dash_n(text):
    return text.replace('/n', ' ')

def remove_spaces(text):
    return re.sub(r"\s+", " ", text).strip()

def remove_lots_of_points(text):
    return re.sub('\.{3,}', ' ', text)

def remove_bad_chars(text):
    return re.sub('[˜˚˝˙ˆˇ˚˘˘˛]', ' ', text)

def remove_numbers(text):
    text = re.sub('-', '', text)
    text = re.sub('\(?R \$ *[0-9\.,]\)?', ' ', text)
    
    text = re.sub('[Nn§]? [§º°]?\.? *[0-9]+[nN§º°\-0-9\.\,\/\\ ][º°]?', ' ', text)
    text = BeautifulSoup(text, 'html.parser') 
    text = re.sub('\W+',' ', text.text)
    return str(text)

def preprocess(text):
    text = remove_dash_n(text)
    text = remove_numbers(text)
    text = remove_bad_chars(text)
    text = remove_lots_of_points(text)
    text = remove_spaces(text)
    return text

def break_paragraphs(text: str):
    return re.split('(?<=\.) *(?=[A-Z])', preprocess(page))

x_train = []
x_train_bert = []
for page in corpus:
  for d in break_paragraphs(page):
      x_train.append(d)
      print(d)
      list_tokens = my_tokenizer(d)
      print(len(list_tokens))
      print(list_tokens)

      # Testes só com BERT
      #print("Bert tokenizer")
      #print(len(bert_tokenizer.tokenize(d)))
      #print(bert_tokenizer.tokenize(d))
      #print(bert_tokenizer.convert_tokens_to_ids(bert_tokenizer.tokenize(d)))
      #print("Bertimbau tokenizer")
      #print(len(bertinbau_tokenizer.tokenize(d)))
      #print(bertinbau_tokenizer.tokenize(d))
      #print(bertinbau_tokenizer.convert_tokens_to_ids(bertinbau_tokenizer.tokenize(d)))
      
```
